```python
import geoviews as gv
import geoviews.feature as gf
import xarray as xr
from cartopy import crs
```

```python
gv.extension('bokeh', 'matplotlib')
```

```python
df = pd.read_excel('point.xlsx')
df
```

```python
points = gv.Points(df, ['Lon', 'Lat'])
points.options(color='k')
```

```python
dataset = gv.Dataset(xr.open_dataset(nc_path))
spam = dataset.select(lon=(148, 160), lat=(20, 27))
spam
```

```python
%%output backend='bokeh'
hoge = spam.to(gv.Image, ['lon', 'lat']).redim.range(chl_ocx=(0.03, 0.08))
hoge.options(cmap='viridis', colorbar=True, width=400) * gf.coastline() 
```

```python
egg = dataset.select(lon=(130, 179), lat=(15, 40))
hage = egg.to(gv.Image, ['lon', 'lat']).redim.range(chl_ocx=(0.03, 0.3))
hage.options(cmap='viridis', colorbar=True, width=400) * gf.coastline() * points.options(color='white', size=5)
```
